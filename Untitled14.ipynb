### Fibonacci (Naïve Recursive)
- **Input:** Integer `n` (position in sequence)  
- **Output:** `n`-th Fibonacci number  
- **Time Complexity:** Best = O(2^n), Avg = O(2^n), Worst = O(2^n)  
- **Space Complexity:** O(n) (due to recursion depth)  
- **Trade-offs:** Very simple but impractical for large `n` because it grows exponentially and may hit recursion depth limit.  


```python
def fib_naive(n):
    if n <= 1:
        return n
    return fib_naive(n-1) + fib_naive(n-2)
```

```python
import time, random
import matplotlib.pyplot as plt
from memory_profiler import memory_usage

def profile_run(fn, arg_generator, sizes):
    times, mems = [], []
    for n in sizes:
        args = arg_generator(n)
        start = time.time()
        mem, _ = memory_usage((fn, (args,)), retval=True, max_iterations=1)
        end = time.time()
        times.append(end - start)
        mems.append(max(mem))
        print(f"{fn.__name__}, n={n}, time={end-start:.5f}s, mem={max(mem):.2f}MB")
    return times, mems

fib_sizes = [5, 10, 20, 25]
times, mems = profile_run(fib_naive, lambda n: n, fib_sizes)

plt.plot(fib_sizes, times, marker="o")
plt.xlabel("n")
plt.ylabel("Time (s)")
plt.title("Fibonacci Naïve")
plt.show()
```

### Fibonacci (Dynamic Programming)
- **Input:** Integer `n`  
- **Output:** `n`-th Fibonacci number  
- **Time Complexity:** Best = O(n), Avg = O(n), Worst = O(n)  
- **Space Complexity:** O(1)  
- **Trade-offs:** Efficient and scales to large inputs easily, unlike the naïve recursive version.  


```python
def fib_dp(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a+b
    return a
```

```python
fib_dp_sizes = [100, 1000, 5000, 10000]
times, mems = profile_run(fib_dp, lambda n: n, fib_dp_sizes)

plt.plot(fib_dp_sizes, times, marker="o")
plt.xlabel("n")
plt.ylabel("Time (s)")
plt.title("Fibonacci DP")
plt.show()
```

### Merge Sort
- **Input:** Unsorted list of numbers  
- **Output:** Sorted list of numbers  
- **Time Complexity:** Best = O(n log n), Avg = O(n log n), Worst = O(n log n)  
- **Space Complexity:** O(n) (extra arrays for merging)  
- **Trade-offs:** Stable and predictable, but needs extra memory.  


```python
def merge_sort(arr):
    if len(arr) <= 1: return arr
    mid = len(arr)//2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    i=j=0; result=[]
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i]); i+=1
        else:
            result.append(right[j]); j+=1
    result.extend(left[i:]); result.extend(right[j:])
    return result
```

```python
sizes = [100, 500, 1000, 2000]
times, mems = profile_run(merge_sort, lambda n: [random.randint(0,1000) for _ in range(n)], sizes)

plt.plot(sizes, times, marker="o")
plt.xlabel("Input size (n)")
plt.ylabel("Time (s)")
plt.title("Merge Sort")
plt.show()
```

### Quick Sort
- **Input:** Unsorted list of numbers  
- **Output:** Sorted list of numbers  
- **Time Complexity:** Best = O(n log n), Avg = O(n log n), Worst = O(n^2)  
- **Space Complexity:** O(log n) recursion average, O(n) worst-case (extra lists)  
- **Trade-offs:** Very fast in practice, but worst-case if pivots are bad.  


```python
def quick_sort(arr):
    if len(arr) <= 1: return arr
    pivot = arr[len(arr)//2]
    left = [x for x in arr if x < pivot]
    mid = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + mid + quick_sort(right)
```

```python
times, mems = profile_run(quick_sort, lambda n: [random.randint(0,1000) for _ in range(n)], sizes)

plt.plot(sizes, times, marker="o")
plt.xlabel("Input size (n)")
plt.ylabel("Time (s)")
plt.title("Quick Sort")
plt.show()
```

### Insertion Sort
- **Input:** Unsorted list of numbers  
- **Output:** Sorted list of numbers  
- **Time Complexity:** Best = O(n), Avg = O(n^2), Worst = O(n^2)  
- **Space Complexity:** O(1)  
- **Trade-offs:** Works well on small/nearly-sorted arrays, bad on large arrays.  


```python
def insertion_sort(arr):
    a = arr[:]
    for i in range(1,len(a)):
        key = a[i]
        j=i-1
        while j>=0 and a[j]>key:
            a[j+1]=a[j]
            j-=1
        a[j+1]=key
    return a
```

```python
small_sizes = [100, 200, 500]
times, mems = profile_run(insertion_sort, lambda n: [random.randint(0,1000) for _ in range(n)], small_sizes)

plt.plot(small_sizes, times, marker="o")
plt.xlabel("Input size (n)")
plt.ylabel("Time (s)")
plt.title("Insertion Sort")
plt.show()
```

### Bubble Sort
- **Input:** Unsorted list of numbers  
- **Output:** Sorted list of numbers  
- **Time Complexity:** Best = O(n), Avg = O(n^2), Worst = O(n^2)  
- **Space Complexity:** O(1)  
- **Trade-offs:** Simple to implement, but inefficient in practice.  


```python
def bubble_sort(arr):
    a = arr[:]
    for i in range(len(a)):
        for j in range(0,len(a)-i-1):
            if a[j]>a[j+1]:
                a[j],a[j+1]=a[j+1],a[j]
    return a
```

```python
times, mems = profile_run(bubble_sort, lambda n: [random.randint(0,1000) for _ in range(n)], small_sizes)

plt.plot(small_sizes, times, marker="o")
plt.xlabel("Input size (n)")
plt.ylabel("Time (s)")
plt.title("Bubble Sort")
plt.show()
```

### Selection Sort
- **Input:** Unsorted list of numbers  
- **Output:** Sorted list of numbers  
- **Time Complexity:** Best = O(n^2), Avg = O(n^2), Worst = O(n^2)  
- **Space Complexity:** O(1)  
- **Trade-offs:** Always slow, rarely used in real systems.  


```python
def selection_sort(arr):
    a = arr[:]
    for i in range(len(a)):
        min_idx=i
        for j in range(i+1,len(a)):
            if a[j]<a[min_idx]:
                min_idx=j
        a[i],a[min_idx]=a[min_idx],a[i]
    return a
```

```python
times, mems = profile_run(selection_sort, lambda n: [random.randint(0,1000) for _ in range(n)], small_sizes)

plt.plot(small_sizes, times, marker="o")
plt.xlabel("Input size (n)")
plt.ylabel("Time (s)")
plt.title("Selection Sort")
plt.show()
```

### Binary Search
- **Input:** Sorted list + target value  
- **Output:** Index of target (or -1 if not found)  
- **Time Complexity:** Best = O(1), Avg = O(log n), Worst = O(log n)  
- **Space Complexity:** O(1)  
- **Trade-offs:** Extremely fast, but works only on sorted lists.  


```python
def binary_search(arr, target):
    low, high = 0, len(arr)-1
    while low<=high:
        mid=(low+high)//2
        if arr[mid]==target: return mid
        elif arr[mid]<target: low=mid+1
        else: high=mid-1
    return -1
```

```python
sizes = [1000, 10000, 100000, 1000000]
times, mems = profile_run(
    lambda arr: binary_search(list(range(arr)), arr-1),
    lambda n: n, sizes
)

plt.plot(sizes, times, marker="o")
plt.xlabel("Input size (n)")
plt.ylabel("Time (s)")
plt.title("Binary Search")
plt.show()
```

## Visual Insights

- **Fibonacci:** Naïve recursion grows exponentially, hitting recursion depth limits around n ≈ 30. DP version scales linearly and can handle very large n easily.  
- **Sorting Algorithms:** Merge and Quick Sort both show O(n log n) growth. Quick Sort is generally faster, but may degrade in worst cases.  
- **Simple Sorts (Insertion, Bubble, Selection):** Clear O(n^2) growth; they quickly become unusable for large inputs, but Insertion Sort is efficient for small arrays.  
- **Binary Search:** Execution time barely increases even for huge input sizes (logarithmic growth).  


## Reflections

- **Expected vs Observed:** The measured times match theoretical complexities.  
  - Fibonacci naïve clearly shows exponential blow-up.  
  - Merge and Quick Sort behave as O(n log n).  
  - Quadratic sorts (Insertion, Bubble, Selection) become slow after a few thousand elements.  
  - Binary Search remains lightning fast.  
- **Trade-offs in Time vs Space:**  
  - Merge Sort requires extra memory but has predictable performance.  
  - Quick Sort is memory efficient but can degrade to O(n^2).  
  - Fibonacci DP uses constant space, unlike the recursive version.  
- **Recursion Depth Risks:**  
  - Fibonacci (naïve) and Quick Sort may hit Python’s recursion limit for large n.  
  - Merge Sort also uses recursion but usually safe due to log n depth.
